```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Sales")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("Income")
plt.suptitle("My shop")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 1, 10])

plt.bar(x, y, color = "darkred", width = 0.6)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
x = np.random.normal (170, 10, 250)
plt.hist(x)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title = "Four fruits:")
plt.show() 
```
